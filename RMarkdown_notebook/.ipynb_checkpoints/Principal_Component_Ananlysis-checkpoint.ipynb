##### Perform PCA
* scale mean to zero and variables to 1 SD
* center and scale corresponds to mean and SD of the variables that were used for scaling prior to implementing PCA
* rotation : provide PC loadings, each column contain corresponding PC loading vectors. 
* there are general min(n-1, p) informative PC in dataset with n obs and p variables.


```r
 pr.out <- prcomp(USArrests, scale =TRUE)
names(pr.out)
```

```r
pr.out$center
pr.out$scale
pr.out$rotation
```

```r

dim(pr.out$x) # dimension of principal components score matrix  
```

```r
# plot  
biplot(pr.out, scale = 0) 
```

### Interpretation 
* PC1 places approximately equal weight on Assault, Murder, and Rape, but with much less weight on UrbanPop.
* Therefore PC1 corresponds to a measure of overall rates of serious crimes.
* The PC2 places most of its weight on UrbanPop and much less weight on the other three features, hence correspond to urbanization 

```r
# Rotate 
pr.out$rotation = -pr.out$rotation
pr.out$x = -pr.out$x
biplot(pr.out, scale = 0)
```

```r
pr.out$sdev

# variance explained by each PC by squring them 
pr.var <- pr.out$sdev^2
```

```r
pve <- pr.var/sum(pr.var)
pve
```

#### Interpret PVE
* 62% variance explained by PC1
* 24.7% variance explained by PC2 

### Scree plot for PVE

```r
# cumsum 
par(mfrow = c(1,2))
plot(pve, type = "b", xlab = "principal components", ylab = "proportion of variance explained",
    ylim = c(0, 1))
plot(cumsum(pve), type = "b", xlab = "principal components", ylab = "cumulative proportion of variance explained",
    ylim = c(0, 1))
```

### Matrix Completion
* solving optimization problem on centered data matrix X is equivalent to computing the first M principal components of the data. 
* single value decomposition is general algorithm for solving this. 

```r
# df into matrix, after centering and scaling column to mean 0 and var 1 
X <- data.matrix(scale(USArrests))
pcob <- prcomp(X)
summary(pcob)
```

```r
# returns 3 components u d v
sX <- svd(X)
names(sX)
```

```r
sX$v
```

```r
# the matrix u is equivalent to matrix of standardized scores, and SD are in vector d. 
pcob$rotation 
```

```r
t(sX$d * t(sX$u))
```

#### creating data with 20 missing values in USA arrest

```r
nomit <- 20
set.seed(15)
ina <- sample(seq(50), nomit)
inb <- sample(1:4, nomit, replace = TRUE)
Xna <- X
index.na <- cbind(ina, inb)
Xna[index.na] <- NA
```

```r
# Function to perform matrix approximation using SVD
fit.svd <- function(X, M = 1) {
    #  Compute the SVD of the matrix X
   svdob <- svd(X)
    #   Reconstruct the matrix using the first M components
   with(svdob,
       u[, 1:M, drop = FALSE] %*%
       (d[1:M] * t(v[, 1:M, drop = FALSE]))
     )
}
```

```r
# Initialize Xhat as Xna (the matrix with missing values)
Xhat <- Xna

# Compute the column means of the non-missing entries
xbar <- colMeans(Xna, na.rm = TRUE)

# Replace the missing values in Xhat with the corresponding column means
Xhat[index.na] <- xbar[inb]
```

```r
# Threshold for stopping criteria
thresh <- 1e-7

# Initial relative error (set arbitrarily high for the first iteration)
rel_err <- 1

# Iteration counter (starts at 0)
iter <- 0

# Logical matrix indicating where the missing values are in Xna
ismiss <- is.na(Xna)

# Mean squared error (MSE) of non-missing elements in the original matrix, after scaling
mssold <- mean((scale(Xna, xbar, FALSE)[!ismiss])^2)

# Mean squared error (MSE) of non-missing elements in the original matrix (Xna)
mss0 <- mean(Xna[!ismiss]^2)
```

```r
while(rel_err > thresh) {
    # Compute the approximation using the first M components
    Xapp <- fit.svd(Xhat, M = 1)
    
    # Replace the missing values in Xhat with those in the approximation (Xapp)
    Xhat[ismiss] <- Xapp[ismiss]
    
    #  Compute the mean squared error (MSS) for the non-missing entries
    mss <- mean(((Xna - Xapp)[!ismiss])^2)
    
    # Calculate the relative error as the change in MSE divided by the original MSE
    rel_err <- (mssold - mss) / mss0
    
    # Update the old MSE value for the next iteration
    mssold <- mss
    
    # Output the iteration count, current MSE, and relative error
    cat("Iter:", iter, "MSS:", mss, "Rel. Err:", rel_err, "\n")
}
```

```r
cor(Xapp[ismiss], X[ismiss])
```
